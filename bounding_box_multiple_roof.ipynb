```python
from ultralytics import YOLO

# Load the trained YOLOv8 model
model = YOLO('/content/best.pt')

# Make predictions on the image
results = model.predict('/content/WhatsApp Image 2024-10-22 at 15.55.37_70f00cf7.jpg')

# Display the results on the image
results[0].plot()


```

```python
# Accessing the bounding boxes, confidence scores, and class predictions
for result in results:
    boxes = result.boxes.xyxy  # Bounding box coordinates (x1, y1, x2, y2)
    confidences = result.boxes.conf  # Confidence scores
    class_ids = result.boxes.cls  # Class predictions (numeric labels)

    print("Bounding boxes:", boxes)
    print("Confidence scores:", confidences)
    print("Class predictions:", class_ids)

# Save the annotated image
model.predict('/content/WhatsApp Image 2024-10-22 at 15.55.37_70f00cf7.jpg', save=True)
```
